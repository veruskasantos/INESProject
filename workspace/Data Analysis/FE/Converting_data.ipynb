```python
# city = "Recife"
city = "Curitiba"
# city = "Campina Grande"

filename = "../data/output/" + city + "/new_feats_data.csv"

input_data = pd.read_csv(filename, index_col=None, header=0)
input_data.shape
```

```python
# converting categorical variables

input_data['shapeId'] = input_data['shapeId'].astype(str)
input_data['routeFrequency'] = input_data['routeFrequency'].astype(str)
input_data['problem'] = input_data['problem'].astype(str)
input_data['alertSubtype'] = input_data['alertSubtype'].astype(str)
# input_data['alertSubtype'].replace("nan", "-", inplace=True)
input_data['alertType'] = input_data['alertType'].astype(str)
# input_data['alertType'].replace("nan", "-", inplace=True)
input_data['jamBlockType'] = input_data['jamBlockType'].astype(str)
# input_data['jamBlockType'].replace("nan", "-", inplace=True)
input_data['alertIsJamUnifiedAlert'] = (input_data['alertIsJamUnifiedAlert'] == True).astype(int)
input_data['alertInScale'] = (input_data['alertInScale'] == True).astype(int)
input_data['busBunching'] = (input_data['busBunching'] == True).astype(int)
input_data['alertSpeed'] = input_data['alertSpeed'].astype('float64')

input_data['problemSB'] = input_data['problemSB'].astype(str)
input_data['alertSubtypeSB'] = input_data['alertSubtypeSB'].astype(str)
# input_data['alertSubtypeSB'].replace("nan", "-", inplace=True)
input_data['alertTypeSB'] = input_data['alertTypeSB'].astype(str)
# input_data['alertTypeSB'].replace("nan", "-", inplace=True)
input_data['jamBlockTypeSB'] = input_data['jamBlockTypeSB'].astype(str)
# input_data['jamBlockTypeSB'].replace("nan", "-", inplace=True)
input_data['alertIsJamUnifiedAlertSB'] = (input_data['alertIsJamUnifiedAlertSB'] == True).astype(int)
input_data['alertInScaleSB'] = (input_data['alertInScaleSB'] == True).astype(int)
input_data['alertSpeedSB'] = input_data['alertSpeedSB'].astype('float64')

print(input_data.dtypes)
```

```python
# Replacing null values

# changing by shape data
input_data.gpsPointId = np.where(input_data.gpsPointId.isnull(), 0, input_data.gpsPointId)
input_data.gpsLat = np.where(input_data.gpsLat.isnull(), input_data.shapeLat, input_data.gpsLat)
input_data.gpsLon = np.where(input_data.gpsLon.isnull(), input_data.shapeLon, input_data.gpsLon)
input_data.distanceToShapePoint = np.where(input_data.distanceToShapePoint.isnull(), 
                                            0, input_data.distanceToShapePoint)
input_data.alertLatitude = np.where(input_data.alertLatitude.isnull(), input_data.shapeLat, 
                                    input_data.alertLatitude)
input_data.alertLongitude = np.where(input_data.alertLongitude.isnull(), input_data.shapeLon, 
                                    input_data.alertLongitude)
input_data.alertDistanceToClosestShapePoint = np.where(input_data.alertDistanceToClosestShapePoint.isnull(), 
                                            0, input_data.alertDistanceToClosestShapePoint)

input_data.gpsPointIdSB = np.where(input_data.gpsPointIdSB.isnull(), 0, input_data.gpsPointIdSB)
input_data.gpsLatSB = np.where(input_data.gpsLatSB.isnull(), input_data.shapeLatSB, input_data.gpsLatSB)
input_data.gpsLonSB = np.where(input_data.gpsLonSB.isnull(), input_data.shapeLonSB, input_data.gpsLonSB)
input_data.distanceToShapePointSB = np.where(input_data.distanceToShapePointSB.isnull(), 
                                            0, input_data.distanceToShapePointSB)
input_data.alertLatitudeSB = np.where(input_data.alertLatitudeSB.isnull(), input_data.shapeLatSB, 
                                    input_data.alertLatitudeSB)
input_data.alertLongitudeSB = np.where(input_data.alertLongitudeSB.isnull(), input_data.shapeLonSB, 
                                    input_data.alertLongitudeSB)
input_data.alertDistanceToClosestShapePointSB = np.where(input_data.alertDistanceToClosestShapePointSB.isnull(), 
                                            0, input_data.alertDistanceToClosestShapePointSB)

#grouping by weekday and filling NA's with the median of the column
numeric_feats = input_data.dtypes[(input_data.dtypes != "object")].index
input_data[numeric_feats] = input_data.groupby(['WEEKDAY(gps_datetime)'])[numeric_feats].transform(lambda x: x.fillna(x.median()))

input_data.head(5)
```

```python
# Converting int column to float column to receive values between 0-1

for col in data:
    if data[col].dtype == 'int64':
        data[col] = data[col].astype('float64', copy=False)
```

```python
# Get label column and remove it from data
y_bb = data[['busBunching']] #for Bus Bunching / classification
y_hd = data[['headway', 'headwayThreshold']] #for headway / regression

cols = ['headway', 'busBunching', "headwayThreshold"]
data.drop(columns=cols, inplace=True)
```

```python
# Normalize data

cols_len = len(data.columns)

for c in range(cols_len):
    col_values = data.iloc[:,c]
    min_col = min(col_values)
    max_col = max(col_values)
    
    col_values_len = len(col_values)
    
    for i in range(col_values_len):
        if(min_col == max_col):
            data.iat[i,c] = 0.0
        else:
            data.iat[i,c] = (col_values[i] - min_col) / (max_col - min_col)
```
