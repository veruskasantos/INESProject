```python
import pandas as pd
import random
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import joblib
import time
import shap
import matplotlib.pyplot as plt
```

```python
city = 'Curitiba'

data_path = '../../data/output/' + city + '/normalized_data_X_FE.csv'
y_path = '../../data/output/' + city + '/y_FE.csv'

data = pd.read_csv(data_path)
y = pd.read_csv(y_path)
print(data.shape)
print(y.shape)
data.head()
```

```python
# same data of rvm test
# for the experiment with the rvm
data_size = 25000 #[25000, 20000, 15000, 10000, 5000]

data = data[0:data_size]
y = y[0:data_size]
print(data.shape)
print(y.shape)
```

```python
# Making training and test data: 80% Training, 20% Test
random.seed(15) #to get always the same set
train_X, test_X, train_Y, test_Y = train_test_split(data, y, test_size=0.20, random_state=7)
```

```python
del data
del y
#del test_X
#del test_Y
```

```python
# Saving test files
test_path = '../../data/output/' + city
#test_X.to_csv(test_path + "/test_X.csv", index=False)
#test_Y.to_csv(test_path + "/test_Y.csv", index=False)

# Reading same test files to be used in the same experiments
test_X = pd.read_csv(test_path + '/test_X.csv')
test_Y = pd.read_csv(test_path + '/test_Y.csv')
```

### CatBoost

```python
# GridSearch to find best parameter values using the smaller dataset (5%)
random.seed(42)

start = time.time()

model = CatBoostClassifier()

grid = {'learning_rate': [0.01, 0.05, 0.1, 0.5, 0.9],
        'iterations': [10, 100, 1000, 10000, 100000],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}

grid_search_result = model.grid_search(grid, 
                                       X=data, 
                                       y=y, 
                                       plot=True)

end = time.time()
print("Execution time: " + str((end - start)/60) + " min")
```

```python
grid_search_result # Curitiba GRID with 3%
```

```python
grid_search_result # Curitiba GRID with 5%
```

```python
# CatBoost after GridSearch
# Curitiba

# values of the gridSearch (previous step) with 3%
#best_learning_rate = 0.05
#best_iterations = 10000
#best_l2_leaf_reg = 3

# values of the gridSearch (previous step) with 5%
best_learning_rate = 0.1
best_iterations = 10000
best_l2_leaf_reg = 5

start = time.time()

model = CatBoostClassifier(
    custom_loss = ['Accuracy', 'Precision', 'Recall', 'F1'],
    iterations = best_iterations,
    learning_rate = best_learning_rate,
    random_seed = 42,
    l2_leaf_reg = best_l2_leaf_reg,
    logging_level = 'Silent'
)

model.fit(
    train_X, train_Y,
    eval_set = (test_X, test_Y),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True
);

joblib.dump(model, 'Models/Saved_' + city + '_CatBoost_GRID_' + str(data_size) + '.pkl')

end = time.time()
print("Execution time: " + str((end - start)/60) + " min")
```

```python
grid_search_result # CityA
```

```python
# CatBoost after GridSearch
# CityA

# values of the gridSearch (previous step)
best_learning_rate = 0.01
best_iterations = 100000
best_l2_leaf_reg = 1

start = time.time()

model = CatBoostClassifier(
    custom_loss = ['Accuracy', 'Precision', 'Recall', 'F1'],
    iterations = best_iterations,
    learning_rate = best_learning_rate,
    random_seed = 42,
    l2_leaf_reg = best_l2_leaf_reg,
    logging_level = 'Silent'
)

model.fit(
    train_X, train_Y,
    eval_set = (test_X, test_Y),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True
);

joblib.dump(model, 'Models/Saved_' + city + '_CatBoost_GRID_' + str(data_size) + '.pkl')

end = time.time()
print("Execution time: " + str((end - start)/60) + " min")
```

```python
# Curitiba

start = time.time()

model = CatBoostClassifier(
    custom_loss = ['Accuracy', 'Precision', 'Recall', 'F1'],
    iterations = 10000,
    learning_rate = 0.9,
    random_seed = 42,
    l2_leaf_reg = 5,
    depth=8,
    logging_level = 'Silent'
)

model.fit(
    train_X, train_Y,
    eval_set = (test_X, test_Y),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True
);

end = time.time()
print("Execution time: " + str((end - start)/60) + " min")
```

```python
# all data - 100%
# explain the model's predictions using SHAP
start = time.time()

model = joblib.load('Models/Saved_' + city + '_CatBoost_BB_class_all_data.pkl')

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(train_X, check_additivity=False)

end = time.time()
print("Execution time: " + str((end - start)/60) + " min")

shap.summary_plot(shap_values, train_X, plot_type="bar", show=False)
plt.savefig('catboost_shap_values_all_data.pdf', bbox_inches = "tight")
```
